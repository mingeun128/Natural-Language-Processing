```python
def get_text_to_pdf(path):
  with open(path, 'rb') as f:
    pdf = PdfFileReader(f)
    page = pdf.getPage(1)
    text = page.extractText()
    return text
```

```python
path = 'Detection_and_Prevention_of_Sinkhole_Attack_on_AODV_Protocol_in_Mobile_Adhoc_Network.pdf'
paper = get_text_to_pdf(path)
wordcloud = WordCloud().generate(paper)
import matplotlib.pyplot as plt
plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

```python
from nltk.corpus import gutenberg
story = gutenberg.open('melville-moby_dick.txt').read()

porter = nltk.PorterStemmer()
tokens = word_tokenize(story)
porter_process = [porter.stem(t) for t in tokens]

word_count = dict()
for word in porter_process:
  word_count[word] = word_count.get(word, 0) + 1
wordcloud = WordCloud().generate_from_frequencies(word_count)

plt.axis("off")
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
