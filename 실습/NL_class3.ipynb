```python
def get_prior(reviews):# 사전확률 구하는 함수
  comedy_count = 0
  action_count = 0
  reviews_count = 0
  for r in reviews:
    if r[1] == 'action':
      action_count += 1
    elif r[1] == 'comedy':
      comedy_count += 1
    reviews_count += 1
  return action_count/reviews_count,comedy_count/reviews_count
```

```python
def get_words_category(reviews):# Action/Comedy 각각 리스트 형태로
  action_words = []
  comedy_words = []
  for r in reviews:
    if r[1] == 'action':
      action_words.extend(r[0].split(','))
    elif r[1] == 'comedy':
      comedy_words.extend(r[0].split(','))
  return action_words,comedy_words
```

```python
def predict(review, priorA, priorC, wordListA, wordListC):# 장르 구하는 함수
  review_words = review.split(',')
  action_p = 1
  comedy_p = 1

  for w in review_words:
    action_p = action_p * (wordListA.count(w.lower()) / len(wordListA))    # P(w1,w2, .. │ Action)
    comedy_p = comedy_p * (wordListC.count(w.lower()) / len(wordListC))    # P(w1,w2, .. │ Comedy)
  action_p *= priorA    # P(w1,w2, .. │ Action)*P(Action)
  comedy_p *= priorC    # P(w1,w2, .. │ Comedy)*P(Comedy)

  print('['+ review + ']' + '\'s action probability is ' + str(action_p))
  print('['+ review + ']' + '\'s comedy probability is ' + str(comedy_p))

  if action_p > comedy_p:
    result = 'action'
  else:
    result = 'comedy'
  return result
```

```python
reviews = [
('fun,couple,love,love', 'comedy'),
('fast,furious,shoot', 'action'),
('Couple,fly,fast,fun,fun', 'comedy'),
('Furious,shoot,shoot,fun', 'action'),
('Fly,fast,shoot,love', 'action')
]
review1 = 'fast,furious,shoot,love'
review2 = 'fast,fun,fun,love,fun'
review3 = 'fun,fun'

priorA, priorC = get_prior(reviews) #Action과 Comedy의 사전확률
wordListA, wordListC = get_words_category(reviews)

print( review1 + ' belongs to '+ predict(review1, priorA, priorC, wordListA, wordListC))
print()
print( review2 + ' belongs to '+ predict(review2, priorA, priorC, wordListA, wordListC))
print()
print( review3 + ' belongs to '+ predict(review3, priorA, priorC, wordListA, wordListC))
print()
```
